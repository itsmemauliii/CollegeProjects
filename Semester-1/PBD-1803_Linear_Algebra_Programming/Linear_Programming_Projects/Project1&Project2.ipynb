```python
# Simple Linear Algebra: Matrix Operations

import numpy as np

print("--- PBD-1803: Linear Algebra - Simple Matrix Operations ---")

# Define two small matrices
matrix_a = np.array([[1, 2],
                     [3, 4]])

matrix_b = np.array([[5, 6],
                     [7, 8]])

print("\nMatrix A:")
print(matrix_a)

print("\nMatrix B:")
print(matrix_b)

# Matrix Addition
matrix_sum = matrix_a + matrix_b
print("\nMatrix A + B (Element-wise Addition):")
print(matrix_sum)

# Matrix Multiplication (dot product)
# This is a core operation in linear algebra
matrix_product = matrix_a @ matrix_b # Using '@' operator for matrix multiplication
# Alternatively: np.dot(matrix_a, matrix_b)
print("\nMatrix A @ B (Matrix Multiplication):")
print(matrix_product)

# Transpose of a matrix
matrix_a_transpose = matrix_a.T
print("\nTranspose of Matrix A:")
print(matrix_a_transpose)

print("\nThese are foundational operations used extensively in various algorithms like regression and machine learning.")
```

```python
# Simple Linear Programming: Visualizing a Single Constraint

import numpy as np
import matplotlib.pyplot as plt

print("\n--- PBD-1803: Linear Programming - Simple Constraint Visualization ---")

# Define variables for the constraint: 2*x + y <= 10
# Assume x and y are non-negative (common in LP problems)
x_values = np.linspace(0, 10, 400) # x from 0 to 10

# Express y in terms of x for the equality: y = 10 - 2*x
y_equality = 10 - 2 * x_values

# Plot the line representing the equality (boundary of the constraint)
plt.figure(figsize=(7, 6))
plt.plot(x_values, y_equality, color='blue', label='2x + y = 10 (Boundary)')

# Shade the feasible region for the inequality (2*x + y <= 10)
# This means any point below or on the line
plt.fill_between(x_values, 0, y_equality, where=(y_equality >= 0), color='lightblue', alpha=0.5, label='Feasible Region (2x + y <= 10, x>=0, y>=0)')

# Set plot limits and labels
plt.xlim(0, 7)
plt.ylim(0, 12)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Visualization of a Single Linear Inequality Constraint")
plt.axvline(0, color='gray', linestyle='--') # x >= 0
plt.axhline(0, color='gray', linestyle='--') # y >= 0
plt.legend()
plt.grid(True)
plt.show()

print("\nThis visualization shows how a single linear inequality defines a 'feasible region' where solutions can exist within a linear programming problem.")
```
